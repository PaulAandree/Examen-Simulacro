```python
class Node:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None

a = Node('a')
b = Node('b')
c = Node('c')
d = Node('d')
e = Node('e')
f = Node('f')

a.left = b
a.right = c
b.left = d
b.right = e
c.right = f
```

```python
def depth_first_values(root):
    if root is None: return []
    left_values = depth_first_values(root.left)
    right_values = depth_first_values(root.right)
    return [root.val] + left_values + right_values
```

```python
depth_first_values(a)
```

```python
from collections import deque

def breadth_first_values(root):
    if root is None:
        return []
    values = []
    queue = deque([root])
    while len(queue) > 0:
        current = queue.popleft()
        values.append(current.val)
        if current.left is not None:
            queue.append(current.left)
        if current.right is not None:
            queue.append(current.right)
    return values
```

```python
breadth_first_values(a)
```

```python
def breadth_first_values2(root):
    if root is None:
        return []
    values = [root.val]
    left_values = breadth_first_values(root.left)
    right_values = breadth_first_values(root.right)
    return values + left_values + right_values
```

```python
breadth_first_values2(a)
```
